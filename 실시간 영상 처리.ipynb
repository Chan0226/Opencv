```python
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('saved.avi',fourcc,20.0,(640,480))
while video.isOpened():
    ret,frame = video.read();
    if not ret:
        print('error open cam')
        break
    out.write(frame)
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xff == 27: # ESC
        break
    
video.release()
out.release() # release 안해주면 기록이 안됨
cv2.destroyAllWindows()
```

### opencv로 저장한 영상 재생

```python
import cv2
filepath = './yolov5/runs/detect/exp2/saved.avi'
cap = cv2.VideoCapture(filepath)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print('fail open video')
        break
    
    cv2.waitKey(20)    
    cv2.imshow('record video',frame)
    
    if cv2.waitKey(1) & 0xff == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
filepath = './yolov5/runs/detect/exp/saved.avi'
names = [0,filepath]
window_titles = ['real','detected']

cap = [cv2.VideoCapture(i) for i in names]

ret = [None]*len(names)
frames = [None]*len(names)
isOnce = True

while True:
    for i, c in enumerate(cap):
        ret[i],frames[i] = c.read()
    
    # detect.py를 호출한다 한번만
 
    if(isOnce):
        !python ./yolov5/detect.py --weights ./yolov5/runs/train/exp3/weights/best.pt --source saved.avi
        isOnce = False
               
    for i, f in enumerate(frames):
        if ret[i]:
            cv2.imshow(window_titles[i],f)
    if cv2.waitKey(1) & 0xff == 27:  #esc
        break
        
for c in cap:
    c.release()
cv2.destroyAllWindows()
```
