```python
import cv2 
img_file = '01.jpg'
img = cv2.imread(img_file)
title = 'IMG'
x,y = 100,100
```

## opencv 창 이동

```python
import cv2
while True:
    cv2.imshow(title,img)
    cv2.moveWindow(title,x,y)
    key = cv2.waitKey(0) & 0xFF  ## && : 논리연산자 , & : 비트 연산자
    print(key,chr(key))
    if key == ord('h'): # h이면 10 픽셀 좌로 이동
        x -= 10 
    elif key == ord('j'): # j이면 10 픽셀 아래로 이동
        y += 10
    elif key == ord('k'): # k이면  10픽셀 위로 이동
        y -= 10
    elif key == ord ('l'): # l이면 10픽셀 우로 이동
        x += 10
    elif key == ord ('q') or key == 27: # esc 종료
        cv2.destroyAllWindows()
        break;
    
```

## 마우스 이벤트

```python
# 마우스 이벤트 이용 동그라미 그리기
import cv2
title = 'mouse event'
img = cv2.imread('01.jpg')
cv2.imshow(title,img)

# 이벤트 처리 함수는 on~ 시작
def onMouse(event,x,y,flags,param):
    print(event,x,y)
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y), 30,(0,0,0,),-1) # 지름 30인 원에 검은색으로 채운 원을 그린다.
        cv2.imshow(title,img)
cv2.setMouseCallback(title, onMouse) # 마우스 이벤트에 onmouse 이벤트 처리

while True:
    if cv2.waitKey(0) & 0xFF == 27:   # 27 : ESC
        cv2.destroyAllWindows()
        break
```

```python
import cv2
import numpy as np
win_name = 'Trackbar'

img = cv2.imread('01.jpg')
cv2.imshow(win_name,img)

def onChange(x):
    print(x)
    r = cv2.getTrackbarPos('R',win_name)
    g = cv2.getTrackbarPos('G',win_name)
    b = cv2.getTrackbarPos('B',win_name)
    print(r, g, b)
    img[:] = [b,g,r]  # 기존 이미지에 새 픽셀을 적용
    cv2.imshow(win_name,img)
    
# 트랙바 생성
cv2.createTrackbar('R',win_name,255,255, onChange)
cv2.createTrackbar('G',win_name,255,255, onChange)
cv2.createTrackbar('B',win_name,255,255, onChange)

while True:
    if cv2.waitKey(1) & 0xFF == 27: # 27: ESC
        cv2.destroyAllWindows()
        break
```
